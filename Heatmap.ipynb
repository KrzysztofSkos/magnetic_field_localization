# Maksymalny błąd dla X, Y i Z

```python
maxX = 0
maxY = 0
maxZ = 0
maxPos = 0
for row in data:
    if row[6] > maxX:
        maxX = row[6]
    if row[7] > maxY:
        maxY = row[7]
    if row[8] > maxZ:
        maxZ = row[8]
    if row[9] > maxPos:
        maxPos = row[9]
print((maxX, maxY, maxZ, maxPos))
```

```python
meanPos = 0.0
counter = 0
for row in data:
    meanPos += row[9]
    counter += 1
meanPos = meanPos/counter
print(meanPos)
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
normalize = matplotlib.colors.Normalize(vmin=0, vmax=maxPos)
# Rysowanie błędu pozycji dla całego pomiaru
for row in data:
    p = ax.scatter(row[0], row[1], row[2], c=row[9], norm=normalize)
plt.colorbar(p)
```
