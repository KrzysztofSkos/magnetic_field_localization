# Maksymalny błąd dla X, Y i Z

```python
maxX = 0
maxY = 0
maxZ = 0
maxPos = 0
for row in data:
    if row[6] > maxX:
        maxX = row[6]
    if row[7] > maxY:
        maxY = row[7]
    if row[8] > maxZ:
        maxZ = row[8]
    if row[9] > maxPos:
        maxPos = row[9]
print((maxX, maxY, maxZ, maxPos))
```

```python
import math
meanPos = 0.0
counter = 0
row9list = []

for row in data:
    if row[9] is not None and not math.isnan(row[9]):
        row9list.append(row[9])
        counter += 1

print(np.mean(row9list))
# Liczba punktów, które się w ogóle nie policzyły
print(counter)
# Łączna liczba nieudanych prób policzenia w dowolnej próbie: 5320
# Łączna liczba prób 2000 * 10 = 20000
```

```python
#Wykres Pola magnetycznego
current = 135000 # [A]
distance = np.linspace(1, 400, 1000)
flux = 2 * 10**(-7) * 100 * current / distance

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(distance, flux)
ax.vlines([100], -0.1, 2.7, color = 'green', linestyles = 'dashed', label = 'Measurement start')
ax.vlines([200, 300], -0.1, 2.7, color = 'red', linestyles = 'dashed', label = 'Measurement end')
plt.xlabel("Distance from magnet [cm]")
plt.ylabel("Magnetic flux [T]")
ax.grid()
#plt.vlines([71, 171, 271], 0, 400)
```
